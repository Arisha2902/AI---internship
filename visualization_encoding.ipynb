Univariate analysis involves the examination of a single variable. It aims to describe and summarize the characteristics of that variable. This analysis is often conducted using measures of central tendency (e.g., mean, median, mode) and measures of dispersion (e.g., range, variance, standard deviation)

Bivariate analysis involves the examination of the relationship between two variables. It aims to determine how one variable changes in relation to another. This analysis is often conducted using correlation coefficients, scatter plots, and cross-tabulations

Multivariate analysis involves the examination of the relationship between multiple variables simultaneously. It aims to understand how multiple variables interact with each other. This analysis is often conducted using techniques such as multiple regression, factor analysis, and cluster analysis.

```python
#encoding
from sklearn.preprocessing import LabelEncoder
Le = LabelEncoder()
```

```python
df["day"] = Le.fit_transform(df["day"])
```

This line uses the fit_transform method of the LabelEncoder object (Le) to convert the categorical values in the 'day' column into numerical labels. The fit_transform method first learns the unique values in the 'day' column and then transforms them into numerical representations. The result is then assigned back to the 'day' column, overwriting the original categorical values with the numerical ones
