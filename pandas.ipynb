```python
import pandas as pd

# From a list
data = [10, 20, 30]
s1 = pd.Series(data)
print("s1")
print(s1)
print()

# With custom labels
s2 = pd.Series(data, index=['a', 'b', 'c'])
print("s2")
print(s2)
print()

# From a dictionary
grades = {'Math': 90, 'Science': 85}
s3 = pd.Series(grades)
print("s3")
print(s3)
print()

```
